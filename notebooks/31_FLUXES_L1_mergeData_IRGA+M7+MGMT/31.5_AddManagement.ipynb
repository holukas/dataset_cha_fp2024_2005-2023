# Imports

```python
import importlib.metadata
import pandas as pd
import warnings
from datetime import datetime
from pathlib import Path

from diive.core.io.filereader import search_files, MultiDataFileReader
from diive.core.io.files import save_parquet, load_parquet
# pd.options.display.width = None
# pd.options.display.max_columns = None
pd.set_option('display.max_rows', 3000)
pd.set_option('display.max_columns', 3000)
warnings.filterwarnings(action='ignore', category=FutureWarning)
warnings.filterwarnings(action='ignore', category=UserWarning)
dt_string = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
version_diive = importlib.metadata.version("diive")
print(f"diive version: v{version_diive}")
```

# Load IRGA and Meteo7 data from parquet (available vars only)

```python
ec = load_parquet(filepath="31.4_CH-CHA_IRGA_Level-1_eddypro_fluxnet_2005-2023_availableVars_meteo7.parquet")
ec
```

# Load management info from CSV(7 variables)

```python
filename = "21.5_mgmt_full_timestamp.csv"
filepath = Path("../20_MANAGEMENT/") / filename
mgmt = pd.read_csv(filepath)
```

```python
mgmt = mgmt.drop('WD', axis=1)
mgmt = mgmt.set_index('TIMESTAMP_MIDDLE')
mgmt.index = pd.to_datetime(mgmt.index)
mgmt
```

# Add management to IRGA output

```python
df = pd.concat([ec, mgmt], axis=1)
df
```

# Save to file

```python
OUTNAME = "31.6_CH-CHA_IRGA-L1+M7+MGMT_2005-2023"
filepath = save_parquet(filename=OUTNAME, data=df)
df.to_csv(f"{OUTNAME}.csv")
```
