# Imports

```python
import importlib.metadata
import pandas as pd
import warnings
from datetime import datetime
from pathlib import Path

from diive.core.io.filereader import search_files, MultiDataFileReader
from diive.core.io.files import save_parquet, load_parquet
# pd.options.display.width = None
# pd.options.display.max_columns = None
pd.set_option('display.max_rows', 3000)
pd.set_option('display.max_columns', 3000)
warnings.filterwarnings(action='ignore', category=FutureWarning)
warnings.filterwarnings(action='ignore', category=UserWarning)
dt_string = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
version_diive = importlib.metadata.version("diive")
print(f"diive version: v{version_diive}")
```

# Load merged L1 results from parquet (available vars only)

```python
ec = load_parquet(filepath="31.2_CH-CHA_IRGA_Level-1_eddypro_fluxnet_2005-2023_availableVars.parquet")
ec
```

# Load meteo from parquet (7 variables)

```python
filename = "16.1_CH-CHA_meteo7_gapfilled_2005-2023.parquet"
filepath = Path("../10_METEO/") / filename
meteo = load_parquet(filepath=filepath)
meteo
```

# Identify meteo data already in EC L1 output

```python
removecols = [v for v in ec if str(v).endswith("_1_1_1")]
print("The following variables will be removed from EC results:")
print(removecols)
```

# Remove meteo variables from EC L1 output

```python
# Remove meteo data from flux results because this meteo has gaps after flux processing
keepcols = []
[keepcols.append(c) for c in ec.columns if "_1_1_1" not in c]
ec_reduced = ec[keepcols].copy()
ec_reduced
```

# Add meteo to EC L1 output

```python
df = pd.concat([ec_reduced, meteo], axis=1)
df
```

# Save to file

```python
OUTNAME = "31.4_CH-CHA_IRGA_Level-1_eddypro_fluxnet_2005-2023_availableVars_meteo7"
filepath = save_parquet(filename=OUTNAME, data=df)
df.to_csv(f"{OUTNAME}.csv")
```
